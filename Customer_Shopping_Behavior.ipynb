```python
#First import the Library
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# Check the dataset 
df.head()
```

```python
# Check the datatype
df.info()
```

```python
# Check statics in the data
df.describe()
```

```python
# Check the null values
df.isnull().sum()
```

```python
# Replace the null value to the median of value by their category wise
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Check the null value , is it clear or not
df.isnull().sum()
```

```python
# Change the column name in the lower case
df.columns = df.columns.str.lower()
```

```python
# Replace the space into underscore
df.columns = df.columns.str.replace(' ','_')
```

```python
#Change the column name purchase_amount_(usd) to purchase_amount
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# Check the result
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create column purchase_frequency_days
frequency_mapping = {'Fortnightly':14,'Weekly':7,'Monthly':30,'Quarterly':90,'Bi-Weekly':14,'Annually':365,'Every 3 Months':90}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Check discount_applied , promo_code_used both table that are same or not , if same than remove promo_code_used this one
# df[['discount_applied','promo_code_used']].head(10)
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Drop column promo_code_used
df = df.drop('promo_code_used',axis = 1)
```
